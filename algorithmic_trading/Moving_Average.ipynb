```python
import pandas as pd 
import numpy as np 
import pandas_datareader as web
import matplotlib.pyplot as plt 
from datetime import datetime
```

```python
start = datetime(2006, 1, 1)
end = datetime(2021, 1, 1)
df = web.DataReader('AAPL', data_source='yahoo', start=start, end=end)
df
```

```python
%matplotlib inline
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Adj Close'], label='AAPL')
plt.title('Apple adjusted close price')
plt.xlabel('Date')
plt.ylabel('Adj. Close ($)')
plt.legend(loc=0)
plt.show()
```

# Create a simple moving average with a 30-day window

```python
df['30 MA'] = df['Adj Close'].rolling(window=30).mean()
df['100 MA'] = df['Adj Close'].rolling(window=100).mean()

df['100 MA'].plot()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df['Adj Close'], label='Adj. close')
plt.plot(df['30 MA'], label='30 MA')
plt.plot(df['100 MA'], label='100 MA')

plt.title('Adj Close, 30 MA and 100 MA')
plt.xlabel('Date')
plt.ylabel('Adj. Close ($)')
plt.legend(loc=0)
plt.show()
```
