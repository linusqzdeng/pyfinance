```python
import pandas as pd
```

```python
dataset = pd.read_excel('mas_car_00_18_2019_adjusted.xlsx')
```

```python
dataset['SIC'] = dataset['Standard Industry Classification Code'].astype(str)

high_tech_sic = [283, 357, 366, 367, 382, 384, 481, 482, 489, 737, 873]
ht_df = pd.DataFrame()

for sic in high_tech_sic:    
    ht_df = pd.concat([ht_df, dataset[dataset['SIC'].str.startswith(f'{sic}')]])

ht_df.reset_index(inplace=True, drop=True)
```
