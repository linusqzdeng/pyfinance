```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data(Excel)_CEOSAL1.xlsx")
y = df['salary']
x = df['ROE']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
results.params
```

```python
print(results.pvalues)
print(results.tvalues)
```

```python
# another method
pred_val = results.fittedvalues.copy()
true_val = df['salary']
residuals = true_val - pred_val

plt.hist(residuals)
plt.xlabel('Distribution of Residuals')
plt.ylabel('Frequency')
# plt.scatter(df['ROE'], residuals)
plt.show()
```

```python
df['residuals'] = residuals
excessive_sal = df[df['residuals'] > 0]['residuals']
low_sal = df[df['residuals'] < 0]['residuals']

print(excessive_sal.head())

# plt.pie(excessive_sal, low_sal)
# plt.shpw()
```
