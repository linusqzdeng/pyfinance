```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline
```

# read the dataset

```python
df = pd.read_excel('C:/Users/dcmn1/Documents/Learning Files/Cass/Term 1/SMM248 Statistics in Finance/week 5/mortgage.xlsx')
```

# Plotting

```python
# Scatter plot for mdebt and income
x = df['income']
y = df['mdebt']

plt.style.use('ggplot')
plt.scatter(x, y)
plt.xlabel('personal income (billion $)')
plt.ylabel('mortgage debt (billion $)')
plt.show()
```

```python
# Scatter plot for mdebt and mcost
a = df['mcost']
b = df['mdebt']

plt.scatter(a, b)
plt.xlabel('mortgage cost (billion $)')
plt.ylabel('mortgage debt (billion $)')
plt.show()
```

# Estimation

```python
# Regression for income and mcost
# Unestricted model
X = df[['income', 'mcost']]
X = sm.add_constant(X)
Y = df['mdebt']

model = sm.OLS(Y, X)
result = model.fit().summary()
print(result)
```

```python
# Regression for restricted model
B = df['mdebt']

model2 = smf.ols('mdebt~', data=df)
result2 = model2.fit().summary()
print(result2)
```
