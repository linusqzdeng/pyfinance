```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from datetime import date
from pandas_datareader import DataReader as web

plt.style.use('ggplot')
```

```python
start = date(2020, 4, 30)
end = date(2021, 1, 31)
```

```python
df = pd.DataFrame()
df = web('UBER', data_source='yahoo', start=start, end=end)

df.head()
```

```python
plt.plot(df.loc['2020-06-30':'2020-12-01']['Close'])
plt.title('Uber\'s close price')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.xticks(rotation=30)
plt.show()
```
