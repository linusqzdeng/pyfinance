```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## One-dimension plot

```python
%matplotlib inline

plt.style.use('ggplot')

np.random.seed(1000)
y = np.random.standard_normal(10)
plt.plot(y.cumsum())
```

## Two-dimension plot

```python
# Generate data
np.random.seed(2000)
y = np.random.standard_normal((20, 2)).cumsum(axis=0)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(y, lw=2.0)
plt.plot(y, 'ro')
```

```python
# legend

plt.figure(figsize=(8, 4))
plt.plot(y[:, 0], label='1st')
plt.plot(y[:, 1], label='2nd')
plt.legend(loc=0)
plt.show()
```

```python
# Subplots: different y axis

y[:, 0] = y[:, 0] * 100
plt.figure(figsize=(7, 4))

fig, ax1 = plt.subplots()
plt.plot(y[:, 0], label='1st')
plt.legend(loc=8)
ax2 = ax1.twinx()
plt.plot(y[:, 1], color='g',label='2nd')
plt.legend(loc=0)

plt.show()
```

```python
# Two separate subplots

plt.figure(figsize=(7, 4))
plt.subplot(211)    # plt.subplot(nrow, ncols, index)
plt.plot(y[:, 0], label='1st')
plt.axis('tight')
plt.legend(loc=0)
plt.subplot(212)
plt.plot(y[:, 1], label='2nd', color='b')
plt.legend(loc=0)
plt.show()
```

```python
# Histogram

x = np.random.standard_normal((1000, 2))

plt.figure(figsize=(7, 4))
plt.hist(x, label=['1st', '2nd'], bins=30, stacked=True)
plt.legend(loc=0)
plt.show()
```

```python
# Scatter plot
%matplotlib inline

c = np.random.randint(0, 10, len(y))

plt.scatter(y[:, 0], y[:, 1], c=c, marker='o')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
plt.colorbar()
plt.show()
```
