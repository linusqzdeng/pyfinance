# Candlestick plot by mplfinance

```python
import mplfinance as mpf
import pandas as pd
import numpy as np 
from pandas_datareader import DataReader as web
from datetime import date
```

```python
start = date(2018, 8, 1)
end = date(2019, 7, 31)

df = web('BILI', data_source='yahoo', start=start, end=end)
df.head()
```

```python
mpf.plot(df.loc['2018-08-01':'2018-08-31'], type='candle', volume=True)
```
